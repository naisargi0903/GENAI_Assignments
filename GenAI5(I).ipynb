```python
# Import necessary libraries
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
from IPython.display import display

# Load the pre-trained Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"  # Model ID for Stable Diffusion v1.5
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Move the model to GPU (if available) for faster generation
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = pipe.to(device)

# Define a creative text prompt
prompt = "An underwater cafe where memaids serve in bubble cups,highly vibrant"

# Generate the image
print("Generating image...")
image = pipe(prompt).images[0]

# Save the generated image
output_path = "creative_artwork.png"
image.save(output_path)
print(f"Image saved to {output_path}")

# Display the image
display(image)
```
